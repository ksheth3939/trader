```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
trades = pd.read_csv('trades.csv')
trades
```

```python
stoch = trades[trades['AlgoType'] == 'Stoch']
rsi = trades[trades['AlgoType'] == 'RSI']

rsi_buy = rsi[rsi['Action'] == 'BUY']
rsi_sell = rsi[rsi['Action'] == 'SELL']
stoch_buy = stoch[stoch['Action'] == 'BUY']
stoch_sell = stoch[stoch['Action'] == 'SELL']

stoch_stocks_sold = set(stoch_sell['Ticker'])
rsi_stocks_sold = set(rsi_sell['Ticker'])

print('stoch sold' , len(stoch_stocks_sold))
print('RSI sold' , len(rsi_stocks_sold))
rsi_sell
```

```python
count = stoch.groupby(by = ['Ticker']).count().sort_values(by = ['Datetime'])
count = count[count['Datetime'] % 2 == 0]
count
print('Stoch', sum(count.Datetime))
bought_sold_stoch = set(count.index).intersection(stoch_stocks_sold)
len(bought_sold_stoch)
stoch_pairs = stoch[stoch.Ticker.isin(bought_sold_stoch)]

count = rsi.groupby(by = ['Ticker']).count().sort_values(by = ['Datetime'])
count = count[count['Datetime'] % 2 == 0]
count
print('RSI', sum(count.Datetime))
bought_sold_rsi = set(count.index).intersection(rsi_stocks_sold)
len(bought_sold_rsi)
rsi_pairs = rsi[rsi.Ticker.isin(bought_sold_rsi)]
rsi_pairs
```

```python
stoch_percents = []
for comp in bought_sold_stoch:
    x = stoch_pairs[stoch_pairs['Ticker'] == comp]
    x = x.groupby(by = ['Action']).sum()
    stoch_percents.append((x['Value'][1] - x['Value'][0]) / x['Value'][0])
print('Stoch' ,np.mean(stoch_percents), np.std(stoch_percents), len(stoch_percents))

rsi_percents = []
for comp in bought_sold_rsi:
    x = rsi_pairs[rsi_pairs['Ticker'] == comp]
    x = x.groupby(by = ['Action']).sum()
#     print('asdf', (x['YFSharePrice'][1] - x['YFSharePrice'][0]) / x['YFSharePrice'][0])
    rsi_percents.append((x['YFSharePrice'][1] - x['YFSharePrice'][0]) / x['YFSharePrice'][0])
print('RSI' , np.mean(rsi_percents), np.std(rsi_percents), len(rsi_percents), min(rsi_percents), max(rsi_percents))
```
